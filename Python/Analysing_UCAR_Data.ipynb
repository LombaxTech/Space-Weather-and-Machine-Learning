```python
from netCDF4 import Dataset
import numpy as np
data = Dataset('TEC_DATA/REPRO2013_2014/podTec_C001.2014.120.00.01.0031.G28.01_2013.3520_nc')
# args for time and location
data
data.variables.keys()
# data
```

```python
time = data.variables['time'][:]
TEC = data.variables['TEC'][:]
elevation = data.variables['elevation'][:]
```

```python
# type(time.data)
time.data.size
# TEC.data.size
time.data[0]
time.data[1]
time.data[2]
time.data.size
```

```python
from matplotlib import pyplot as plt
x_values = time.data
y_values = TEC.data
y_values_2 = elevation.data

plt.plot(x_values, y_values, color="c", label="TEC")
plt.plot(x_values, elevation.data, color="b", label="elevation")
plt.xlabel("Time")
plt.ylabel("TEC")
plt.title("TEC vs Time")
plt.grid(True)
plt.legend()
plt.show()
```
